```python
import pandas as pd # type: ignore
import numpy as np # type: ignore
import seaborn as sns # type: ignore
from sqlalchemy import create_engine # type: ignore

engine = create_engine('postgresql://renan@localhost:5432/main')
sql = '''
    select * from flightsearcher.summary_view;
'''
df = pd.read_sql_query(sql,con=engine)
dfFilter = df.copy() #[
    #(df['origin_country'] != 'Usa') &
    #(df['destiny_country'] != 'Usa') &
    #(df['origin_country'] != 'Canada') 
    #(df['destiny_country'] != 'Canada')
#]
dfFilter[['hours','minutes_raw']] = dfFilter.duration.str.split('h',expand=True)
dfFilter['minutes'] = dfFilter.minutes_raw.str.extract('(\d+)')
dfFilter['total_hours'] = dfFilter['hours'].astype(float) + dfFilter['minutes'].astype(float)/60
originSP = dfFilter[dfFilter['origin_city']=='Sao Paulo']
destinySP = dfFilter[dfFilter['destiny_city']=='Sao Paulo']
lane = originSP.merge(destinySP, left_on ='destiny_city', right_on='origin_city',how='inner',suffixes=('_ida','_volta'))
lane['total_price'] = lane['price_ida'] * 0.8 + lane['price_volta']  * 0.8
lane['total_hours'] = lane['total_hours_ida'] + lane['total_hours_volta']
lane['price_per_hour'] = lane['total_price'] / lane['total_hours']
lane['price_for_two'] = lane['total_price'] * 2
lane['12_times_price'] = lane['total_price'] / 12
lane['12_times_price_for_two'] = lane['price_for_two'] / 12
```

```python
lane[
    (lane['destiny_country_ida']=='Spain') &
    (lane['travel_date_volta'] > lane['travel_date_ida'])
].pivot_table(index='travel_date_ida',columns='travel_date_volta',values='price_for_two',aggfunc='min',margins=True).style.background_gradient(cmap='coolwarm',axis=None).format("{:.2f}")
```
